```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = pd.read_csv("dataforviolinPlot.csv")
```

```python
idx = data["Group"] == "Patel's Tau"
data_tau = data[idx]
data_neo = data[-idx]
```

```python
data_neo
```

```python
plt.plot(data_tau["Spec_dist"], data_tau["Data"],  label = "Neuro ODE")
```

```python
plt.plot(data_neo["Spec_dist"], data_neo["Data"],  label = "Neuro ODE")
```

```python
data.groupby('Group').plot(x = 'Data',  y='Spec_dist',legend=True,xlabel="Sim dataset", ylabel="Spec_dist")
```
